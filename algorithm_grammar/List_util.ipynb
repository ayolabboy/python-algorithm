```python
## 리스트 생성 및 출력 
a = [1, 3, 5, 7, 9]
print(a)
print(a[3])
b = [1, 3, 5, 7, [1,2,3]]
print(b)
print(b[4][1])
```

```python
## 리스트 슬라이싱
a = [1, 3, 5, 7, 9]
a[0:3]
```

```python
## 리스트 연산
a = [1,2,3]
b = [4,5,6]
print(a+b)
print(a*2)
```

```python
## 리스트 길이
a = [1,2,3]
print(len(a))
```

```python
## 리스트 요소 삭제
a =[1,2,3,4]
del(a[2:])
print(a)
```

```python
## 리스트 추가
a = [1,2,3,4]
a.append(5)
print(a)
```

```python
## 리스트 삽입
a = [1, 2, 3]
a.insert(0, 4)
print(a)
```

```python
## 리스트 제거
a = [1, 2, 3]
a.pop()
print(a)
```

```python
a = [1, 2, 3, 1, 2, 3]
a.remove(3)
print(a)
```

```python
## 리스트 정렬 
a = [1,4,5,2]
a.sort()
print(a)
```

```python
## 리스트 카운트
a = [1,2,3,1]
cnt = a.count(1)
print(cnt)
```

```python
## 리스트 익스텐션 
a = [1,2,3]
a.extend([4,5])
print(a)
b = [6, 7]
a.extend(b)
print(a)
```
